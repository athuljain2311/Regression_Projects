```python
player_data['HighestScore'] = player_data['HighestScore'].apply(lambda x: 0 if x == 'DNB' else x).astype(int)
```

```python
player_data['IsDismissed'] = player_data['IsDismissed'].apply(lambda x: 'NA' if x is np.nan else x)
```

```python
bat_df = player_data[(player_data['Year'] == year) & (player_data['HighestScore'] > 0)][['Name','HighestScore','IsDismissed']]
```

```python
bat_df = bat_df.sort_values(['HighestScore','IsDismissed'],ascending=[False,False])
```

```python
bat_df['HighestRuns'] = bat_df['HighestScore'].astype(str) + bat_df['IsDismissed'].apply(lambda x: '*' if x == 'Not Dismissed' else '')
```

```python
bat_df[['Name','HighestRuns']].head(50)
```

```python
bowl_df = player_data[(player_data['Year'] == year) & (player_data['BestBowlingWickets'] > 0)][['Name','BestBowlingWickets','BestBowlingRuns']].sort_values(by=['BestBowlingWickets','BestBowlingRuns'],ascending=[False,True])
```

```python
bowl_df['BestFigures'] = bowl_df['BestBowlingWickets'].astype(str) + '/' + bowl_df['BestBowlingRuns'].astype(str)
```

```python
teams = list(player_data[player_data['Year'] == year]['TeamName'].unique())
```

```python
age_data = [player_data[player_data['Year'] == year]['Age']]
for team in teams:
    age_data.append(player_data[(player_data['Year'] == year) & (player_data['TeamName'] == team)]['Age'])
```

```python
teams.insert(0,'Overall Age')
```

```python
fig = ff.create_distplot(age_data,teams,show_hist=False,show_rug=False)
fig.show()
```

```python
batting_age = player_data[(player_data['Year'] == year) & (player_data['TotalRuns'] >= 200)]['Age']
```

```python
bowling_age = player_data[(player_data['Year'] == year) & (player_data['Wickets'] >= 10)]['Age']
```

```python
fig = ff.create_distplot([batting_age,bowling_age],['Run Scorers','Wicket Takers'],show_hist=False,show_rug=False)
fig.show()
```

```python
import plotly.express as px
```

```python
runs = player_data[player_data['Year'] == year][['Nationality','TotalRuns']].groupby('Nationality').sum().reset_index()
runs.rename(columns={'Nationality':'Domicile','TotalRuns':'Runs Scored'},inplace=True)
runs
```

```python
fig = px.pie(runs,values='Runs Scored',names='Domicile')
fig.show()
```

```python
player_count = player_data[player_data['Year'] == year]['Nationality'].value_counts().reset_index()
player_count.rename(columns={'index':'Domicile','Nationality':'Strength'},inplace=True)
player_count
```

```python
fig = px.pie(player_count,values='Strength',names='Domicile')
fig.show()
```
