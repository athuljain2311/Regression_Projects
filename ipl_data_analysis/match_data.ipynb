```python
def match_name_splitter(match_name):
    match_name = match_name.lower()
    first,second = match_name.split('vs')
    return first,second
```

```python
def comments_parser(comment):
    comment = comment.lower()
    if 'won' in comment:
        if 'tied' not in comment:
            return comment[:comment.find('won')],comment[comment.find('won'):]
        else:
            open = comment.find('(')
            close = comment.find('won')
            return comment[open+1:close],'Match tied( won the Super Over)'
    else:
        return '',comment.capitalize()
```

```python
for df in matches.values():
    df['MatchName1'] = df.MatchName.apply(lambda x: match_name_splitter(x)[0].title())
    df['MatchName2'] = df.MatchName.apply(lambda x: match_name_splitter(x)[1].title())
    df['Comments1'] = df.Comments.apply(lambda x: comments_parser(x)[0].title())
    df['Comments2'] = df.Comments.apply(lambda x: comments_parser(x)[1])
    
    df.drop(columns=['MatchName','Comments','Winner'],inplace=True)
```

```python
final_df = None

for df in matches.values():
    if final_df is None:
        final_df = df
    else:
        final_df = pd.concat([final_df,df],ignore_index=True)
```

```python
final_df['1Summary'].fillna('NA',inplace=True)
final_df['2Summary'].fillna('NA',inplace=True)
```

```python
def summary_parser(summary):
    if summary != 'NA':
        score = summary.split()[0]
        runs,wickets = score.split('/')
        overs = summary.split()[1][1:]
        return runs,wickets,overs
    else:
        return 'NA','NA','NA'
```

```python
final_df['Runs1'] = final_df['1Summary'].apply(lambda x: summary_parser(x)[0])
final_df['Wickets1'] = final_df['1Summary'].apply(lambda x: summary_parser(x)[1])
final_df['Overs1'] = final_df['1Summary'].apply(lambda x: summary_parser(x)[2])

final_df['Runs2'] = final_df['2Summary'].apply(lambda x: summary_parser(x)[0])
final_df['Wickets2'] = final_df['2Summary'].apply(lambda x: summary_parser(x)[1])
final_df['Overs2'] = final_df['2Summary'].apply(lambda x: summary_parser(x)[2])
```

```python
final_df['RevisedOver'].fillna(0,inplace=True)
final_df['RevisedTarget'].fillna(0,inplace=True)
```
