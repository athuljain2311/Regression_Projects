```python
dfs = {}
for year in range(2008,2024):
    df = pd.read_csv(f'runs_wickets_per_season\\{year}_batters.csv')
    if year in [2022,2023]:
        df.drop(columns=['ClientPlayerID'],inplace=True)
    else:
        df.drop(columns=['Most_4s','Most_6s'],inplace=True)
    dfs[year] = df
    print(f'{year} : {len(df.columns)}, {len(set(df.columns))}')
```

```python
cols, year, flag = None, 2008, 0

for df in dfs.values():
    if cols is None:
        cols = sorted(df.columns)
    else:
        if cols != sorted(df.columns):
            print(year)
            flag = flag + 1
    year = year + 1

if flag == 0:
    print('All Dataframes have the same Columns')
```

```python
from bs4 import BeautifulSoup
from googlesearch import search
```

```python
def googler(query):
    for url in search(query,num=10):
        if 'www.espncricinfo.com' in url:
            return url
```

```python
espn_url = googler('Yo Mahesh')
```

```python
print(espn_url)
```

```python
import requests
```

```python
response = requests.get(espn_url)
```

```python
soup = BeautifulSoup(response.content,'lxml')
```

```python
for data in soup.find_all('span',class_='ds-text-comfortable-s'):
    print(data.text)
```

```python
count = 0
lst = {}
temp_lst = []
for data in soup.find('div',class_='ds-grid lg:ds-grid-cols-3 ds-grid-cols-2 ds-gap-4 ds-mb-8').find_all('p'):
    count = count + 1
    temp_lst.append(data.text)
    if count == 2:
        lst[temp_lst[0]] = temp_lst[1]
        count = 0
        temp_lst = []
print(lst)
```
