```python
def cumulative(name,auction_year,new_df):
    gamma = 0.2
    
    innings_sum = 0
    runs_scored_sum = 0
    balls_faced_sum = 0
    outs_sum = 0
    wickets_sum = 0
    balls_bowled_sum = 0
    runs_given_sum = 0
    nationality = 'Not Available'
    age = 0
    
    try:
        for year in range(2008,auction_year):
            df = player_data[(player_data['Name'] == name) & (player_data['Year'] == year)]

            if df.empty:
                continue
            else:
                data = list(df.loc[:,['Innings','TotalRuns','Balls','Outs','Wickets','BallsBowled','TotalRunsConceded']].values[0])

                penalty = np.exp(-gamma*(auction_year-year))

                innings = data[0]
                runs_scored = data[1]
                balls_faced = data[2]
                outs = data[3]
                wickets = data[4]
                balls_bowled = data[5]
                runs_given = data[6]

                innings_sum = innings_sum + innings
                runs_scored_sum = runs_scored_sum + penalty*runs_scored
                balls_faced_sum = balls_faced_sum + penalty*balls_faced
                outs_sum = outs_sum + penalty*outs
                wickets_sum = wickets_sum + penalty*wickets
                balls_bowled_sum = balls_bowled_sum + penalty*balls_bowled
                runs_given_sum = runs_given_sum + penalty*runs_given
            
        nationality = player_data[(player_data['Name'] == name) & (player_data['Year'] == auction_year)]['Nationality'].values[0]
        age = player_data[(player_data['Name'] == name) & (player_data['Year'] == auction_year)]['Age'].values[0]
        
        data_dict = {'Name':[name],'Year':[auction_year],'Nationality':[nationality],'Age':[age],'Innings':[innings_sum],'TotalRuns':[runs_scored_sum],'Outs':[outs_sum],'Wickets':[wickets_sum],'BallsBowled':[balls_bowled_sum],'TotalRunsConceded':[runs_given_sum]}
        
    except Exception as e:
        print(name,e)
        data_dict = {'Name':[name],'Year':[auction_year],'Nationality':[nationality],'Age':[age],'Innings':[innings_sum],'TotalRuns':[runs_scored_sum],'Outs':[outs_sum],'Wickets':[wickets_sum],'BallsBowled':[balls_bowled_sum],'TotalRunsConceded':[runs_given_sum]}
    
    if new_df is None:
        new_df = pd.DataFrame(data_dict)
    else:
        new_df = pd.concat([new_df,pd.DataFrame(data_dict)])
        
    return new_df
```

```python
new_df = None
for year in range(2013,2024):
    for name in auction[year]['PLAYER']:
        new_df = cumulative(name,year,new_df)
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
x_train.columns
```

```python
cat_columns = ['Nationality']
num_columns = ['Age','Innings','TotalRuns','Outs','Wickets','BallsBowled','TotalRunsConceded']
```

```python
num_pipeline = Pipeline(steps=[
    ('scaler',StandardScaler(with_mean=False))
])
```

```python
cat_pipeline = Pipeline(steps=[
    ('ohe',OneHotEncoder()),
    ('scaler',StandardScaler(with_mean=False))
])
```

```python
preprocessor = ColumnTransformer([
    ('categorical_pipeline',cat_pipeline,cat_columns),
    ('numerical_pipeline',num_pipeline,num_columns)
])
```

```python
x_train = preprocessor.fit_transform(x_train)
```

```python
x_test = preprocessor.transform(x_test)
```
