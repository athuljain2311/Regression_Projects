```python
def highest_score_parser(score):
    if not score is np.nan:
        if score.endswith('*'):
            return score[:-1],'Not Dismissed'
        else:
            return score,'Dismissed'
    else:
        return 'DNB','NA'
```

```python
def bowling_parser(figures):
    if figures is np.nan:
        return 0,0
    else:
        wickets,runs = tuple(map(int,figures.split('/')))
        return wickets,runs
```

```python
for year,df in combined.items():
    df['Year'] = year
    df['Age'] = year - df['PlayerDOB'].apply(lambda x: x.split('-')[-1]).astype(int)
```

```python
for df in combined.values():
    for column in df.columns:
        if df[column].dtype == 'object' or df[column].dtype == 'bool':
            continue
        else:
            df[column].fillna(0,inplace=True)
```

```python
for df in combined.values():
    df['IsDismissed'] = df.HighestScore.apply(lambda x:highest_score_parser(x)[1])
    df['HighestScore'] = df.HighestScore.apply(lambda x:highest_score_parser(x)[0])
    df.drop(columns=['BBMW'],inplace=True)
```

```python
for df in combined.values():
    df['BestBowlingWickets'] = df.BBIW.apply(lambda x:bowling_parser(x)[0])
    df['BestBowlingRuns'] = df.BBIW.apply(lambda x:bowling_parser(x)[1])
    df.drop(columns=['BBIW'],inplace=True)
```
