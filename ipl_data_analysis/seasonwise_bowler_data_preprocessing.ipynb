```python
def number_of_balls(overs):
    if '.' in overs:
        over,ball = list(map(int,overs.split('.')))
    else:
        over,ball = int(overs),0
    return over*6 + ball
```

```python
dfs[2008]['OversBowled']
```

```python
dfs[2008]['BallsBowled'] = dfs[2008]['OversBowled'].apply(number_of_balls)
```
