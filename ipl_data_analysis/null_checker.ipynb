```python
def ground_name(name):
    grounds = {
        'Brabourne - CCI' : 'Brabourne Stadium',
        'Dubai International Stadium' : 'Dubai International Cricket Stadium',
        'HPCA Stadium' : 'Himachal Pradesh Cricket Association Stadium',
        'MA Chidambaram Stadium' : 'M. A. Chidambaram Stadium',
        'OUTsurance Oval' : 'Mangaung Oval',
        'IS Bindra Stadium' : 'Punjab Cricket Association IS Bindra Stadium',
        'Rajiv Gandhi Intl. Cricket Stadium' : 'Rajiv Gandhi International Stadium'
    }
    
    if name not in grounds.keys():
        return name
    else:
        return grounds[name]
```

```python
matches['GroundName'] = matches['GroundName'].apply(ground_name)
```

## Removing null values from player_data

```python
player_stats = pd.read_csv('seasonwise/for_all_seasons_2.csv')
```

```python
player_stats.isnull().sum()
```

```python
player_stats[player_stats['BattingAverage'].isnull()][['Name','TotalRuns']]
```

```python
player_stats['BattingAverage'].fillna(-1,inplace=True)
```

```python
player_stats[(player_stats['BestScore'].isnull()) & (player_stats['TotalRuns'] == 0)]
```

```python
# player_stats.to_csv('seasonwise/player_stats_all_time.csv')
```

```python
player_stats.info()
```

```python
object_col = []
for column in player_stats.columns:
    if player_stats[column].dtype == 'object':
        object_col.append(column)
```

```python
object_col
```

```python
player_stats[player_stats['BattingAverage'].apply(lambda x:type(x)==float)][['Name','BattingAverage']]
```

```python
player_stats['BattingAverage'] = player_stats['BattingAverage'].apply(lambda x: -1 if x == '-' else x).astype(float)
```

```python
player_stats[player_stats['BestScore'].apply(lambda x:type(x)==str)][['Name','BestScore']]
```

```python
player_stats['BestScore'] = player_stats['BestScore'].apply(lambda x:'Not Available' if x is np.nan else x)
```

```python
num_col = []
for column in player_stats.columns:
    if player_stats[column].dtype != 'object':
        num_col.append(column)
```

```python
num_col
```

```python
player_stats[player_stats['IsNotDismissed'] == -1][['Name','TotalRuns']].sort_values('TotalRuns',ascending=False)
```

```python
player_stats[(player_stats['Outs'] == 0) & (player_stats['TotalRuns'] == 0)][['Name','TotalRuns','Outs','BattingAverage','HighestScore','BestScore']]
```

```python
player_stats['BattingAverage'][player_stats['Outs'] == 0] = -1
```

```python
player_stats[(player_stats['Wickets'] == 0) & (player_stats['BallsBowled'] > 0)][['Name','BowlingAverage','Wickets']]
```

```python
player_stats.to_csv('seasonwise/player_stats_all_time.csv',index=False)
```
