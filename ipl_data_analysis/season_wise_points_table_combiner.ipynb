```python
import pandas as pd
import numpy as np
```

```python
points_table = {}
```

```python
for year in range(2008,2024):
    points_table[year] = pd.read_csv(f'seasonwise/points_table/table_{year}.csv')
```

```python
for df in points_table.values():
    df.drop(columns=['TeamCode'],inplace=True)
```

```python
for year,df in points_table.items():
    number_of_teams = df.shape[0]
    df['Standings'] = pd.Series(range(1,number_of_teams+1))
    df['Year'] = year
```

```python
final_df = None
```

```python
for df in points_table.values():
    if final_df is None:
        final_df = df
    else:
        final_df = pd.concat([final_df,df],ignore_index=True)
```

```python
final_df.sort_values(by=['Year','Standings'],ascending=[False,True])
```

```python
final_df.to_csv('seasonwise/points_table_all_seasons.csv',index=False)
```
